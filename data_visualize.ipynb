# 데이터 시각화

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

test1_all = pd.read_csv("../../data/test1_all.csv")


print(test1_all.shape)
print(test1_all.columns)
```

```python
# 데이터 연령 분포
sns.violinplot(y="age", data=test1_all)
print("10세 미만 수 : ", ((test1_all["age"] < 10)).sum())
print("10대 수 : ", ((test1_all["age"] >= 10) & (test1_all["age"] < 20)).sum())
print("20대 수 : ", ((test1_all["age"] >= 20) & (test1_all["age"] < 30)).sum())
print("30대 수 : ", ((test1_all["age"] >= 30) & (test1_all["age"] < 40)).sum())
print("40대 수 : ", ((test1_all["age"] >= 40) & (test1_all["age"] < 50)).sum())
print("50대 수 : ", ((test1_all["age"] >= 50) & (test1_all["age"] < 60)).sum())
print("60대 수 : ", ((test1_all["age"] >= 60) & (test1_all["age"] < 70)).sum())
print("70대 수 : ", ((test1_all["age"] >= 70) & (test1_all["age"] < 80)).sum())
print("80세 이상 수 : ", ((test1_all["age"] >= 80)).sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap
plt.figure(figsize=(14,10))
sns.heatmap(test1_all.corr()[["PHQ_score"]].sort_values("PHQ_score", ascending=False),
            annot=True, cmap="coolwarm")

# PHQ_score와의 상관만 추출
corr_PHQ = test1_all.corr()["PHQ_score"]
corr_PHQ_sorted = corr_PHQ.sort_values(ascending=False)
print(corr_PHQ_sorted)
```
