# 6. 문자 -> 숫자 인코딩

```python
import pandas as pd

# 문자열(범주형) 데이터 -> 숫자로 변환

# ID 컬럼 제거 (고유 식별자로 분석에 불필요)
if 'ID' in test1_all.columns:
    test1_all.drop(['ID'], axis=1, inplace=True)
    print("ID 컬럼 제거 완료.")

# 문자열(Object) 타입 컬럼 확인
# ID를 제외한 후, 남아있는 object 타입을 찾기
object_cols = test1_all.select_dtypes(include=['object']).columns
print(f"\n인코딩할 문자열 컬럼: {list(object_cols)}")

if len(object_cols) > 0:
    # 나머지 문자열 컬럼을 원-핫 인코딩으로 변환
    # pd.get_dummies)범주형 데이터 -> 숫자
    test1_all = pd.get_dummies(test1_all, columns=object_cols, drop_first=True)
    print("One-Hot Encoding 완료.")
else:
    print("ID를 제외하고 남아있는 문자열 컬럼이 없습니다. 인코딩 건너뜀.")
```
