# 5. 결측치 처리
- 9, 99, 999(무응답) -> 결측치 처리

```python
na_9 = [
    'BD1', 'BD1_11', 'BD2_1', 'BD2_31', 'BD7_4', 'BD7_5',
    'BE3_71', 'BE3_72', 'BE3_81', 'BE3_82',
    'BE8_1', 'BE8_2',
    'BO1', 'BO1_1', 'BO1_2', 'BO1_3', 'BO2_1',
    'BP1', 'BP5', 'BP6_10', 'BP6_2', 'BP6_31', 'BP7',
    'BS10_1', 'BS12_1', 'BS12_2', 'BS13',
    'BS8_2', 'BS9_2',
    'CH2_1',
    'EC1_1', 'EC_lgw_2', 'EC_lgw_4', 'EC_lgw_5', 'EC_occp',
    'EC_pedu_1', 'EC_pedu_2', 'EC_stt_1', 'EC_wht_0', 'EC_wht_23',
    'LQ2_ab', 'LQ4_00',
    'LQ_1EQL', 'LQ_2EQL', 'LQ_3EQL', 'LQ_4EQL', 'LQ_5EQL'
]

na_99 = [
    'BE3_31', 'BE3_32', 'BE3_33',
    'BE3_73', 'BE3_74',
    'BE3_75', 'BE3_76', 'BE3_77', 'BE3_78',
    'BE3_83', 'BE3_84', 'BE3_85', 'BE3_86', 'BE3_87', 'BE3_88',
    'BE3_91', 'BE3_92', 'BE3_93', 'BE3_94',

    'BS10_2',
    'BS3_1', 'BS3_2', 'BS6_3',

    'D_1_1', 'D_2_1', 'D_2_wk',

    'LQ1_mn', 'LQ1_sb', 'LQ2_mn'
]


na_999 = [
    'BD2',         # 음주 시작 연령
    'BS10_3',      # 흡연량
    'BS2'          # 몇몇 흡연 숫자형 변수
]


def replace_missing(df):
    df = df.copy()

    df[na_9]    = df[na_9].replace(9,    np.nan)
    df[na_99]   = df[na_99].replace(99,   np.nan)
    df[na_999]  = df[na_999].replace(999,  np.nan)
    return df


test1_all = replace_missing(test1_all)

# 최빈값으로 결측치 대체
for col in test1_all.columns:
    if test1_all[col].isna().sum() > 0:
        mode_val = test1_all[col].mode()[0]
        test1_all[col] = test1_all[col].fillna(mode_val)

# 결측치 확인
missing = test1_all.isna().sum()[test1_all.isna().sum() > 0]
print("남은 결측치 열 개수:", missing.shape[0])
```

# 6. 문자 -> 숫자 인코딩

```python
import pandas as pd

# 문자열(범주형) 데이터 -> 숫자로 변환

# ID 컬럼 제거 (고유 식별자로 분석에 불필요)
if 'ID' in test1_all.columns:
    test1_all.drop(['ID'], axis=1, inplace=True)
    print("ID 컬럼 제거 완료.")

# 문자열(Object) 타입 컬럼 확인
# ID를 제외한 후, 남아있는 object 타입을 찾기
object_cols = test1_all.select_dtypes(include=['object']).columns
print(f"\n인코딩할 문자열 컬럼: {list(object_cols)}")

if len(object_cols) > 0:
    # 나머지 문자열 컬럼을 원-핫 인코딩으로 변환
    # pd.get_dummies)범주형 데이터 -> 숫자
    test1_all = pd.get_dummies(test1_all, columns=object_cols, drop_first=True)
    print("One-Hot Encoding 완료.")
else:
    print("ID를 제외하고 남아있는 문자열 컬럼이 없습니다. 인코딩 건너뜀.")
```
