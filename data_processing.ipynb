# 3. 다른 연도 (2022 ~ 2014) 와 공통 열 찾고 모든 데이터 합치기


```python
test_dic[2014] = test_dic[2014].rename(columns={"id": "ID"})

target = "ID"

for year, df in test_dic.items():
    cols = [col for col in df.columns if col.startswith(target)]
    print(year, cols)
```

```python
# 1) 공통열 찾기
dfs = list(test_dic.values())

common_cols = set(dfs[0].columns)
for df in dfs[1:]:
    common_cols = common_cols.intersection(df.columns)

common_cols = sorted(list(common_cols))
print(len(common_cols))


bp_cols = [col for col in common_cols if col.startswith("BP")]
print(bp_cols)


# 2) 데이터 합치기
test1_all = pd.concat([df[common_cols] for df in dfs], ignore_index=True)
print(test1_all.shape)
```

# 4. 주관식, 기타(생성변수) 제거
- "기타상세" -> 서술형에 해당하는 열 drop
- `"occp", "edu"`: 최빈값
- `"ID_F", "ID_M"` : 값이 있으면 1, NaN 이면 0



```python
test1_all = pd.concat([df[common_cols] for df in dfs], ignore_index=True)

# 주관식, 기타 제거
test1_all.drop(["ID_fam", "year", "psu", "LQ4_24", "age_month", "incm5", "ho_incm5", 
                "mh_PHQ_S", "EQ5D", "pa_aerobic","sm_presnt", "mh_stress", "dr_month", 
                "wt_tot", "wt_ntr", "ainc", "wt_itvex", "BS6_4", "mod_d", "wt_hs", 
                "kstrata", "genertn", "ainc_unit1", "ainc_1", "tins", "npins",
                 "LQ4_01", "LQ4_02", "LQ4_03", "LQ4_04", "LQ4_05", "LQ4_06", "LQ4_07", "LQ4_08",
                  "LQ4_09", "LQ4_10", "LQ4_11", "LQ4_12", "LQ4_13", "LQ4_14", "LQ4_15", "LQ4_16", "LQ4_21", 
                  "LQ4_22", "LQ4_23", "CH2_2", "EC_stt_2", "BD2_32", "BS3_3", "BS6_2", "BS6_2_1", "BS6_2_2", 
                  "BS6_4", "BS6_4_1", "BS6_4_2", "BS5_1", ], axis = 1, inplace = True)

# 결과
missing = test1_all.isna().sum()[test1_all.isna().sum() > 0]
print("결측치가 있는 열 개수 :", missing.shape[0])
print(missing)
```

# 5. 결측치 처리

```python
# 결측치 처리
for col in test1_all.columns:
    if test1_all[col].isna().sum() > 0:      # 결측치가 있는 열만
        m = test1_all[col].mode()[0]  # 최빈값 추출
        test1_all[col] = test1_all[col].fillna(m)

# 결과
missing = test1_all.isna().sum()[test1_all.isna().sum() > 0]
print("결측치가 있는 열 개수 :", missing.shape[0])
```
